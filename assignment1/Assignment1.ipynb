## Task 2. Implement [Byte-Pair Encoding(BPE)](https://arxiv.org/pdf/1508.07909.pdf) Tokenizer (80 points)

### Task 2.1. Implementation (60 points)

Implement the tokenizer as the BPETokenizer class:
* Implement `train` method that learns merges and builds the vocabulary of the specified `vocab_size` (25 points).
* Implement `tokenize` method that should tokenize the text according to the learnt merges (25 points).

Your code should have docstrings and comments (10 points).

```python
import re
import collections
from typing import List

class BPETokenizer:
    def __init__(self, vocab_size: int) -> None:
        """Initialize the BPETokenizer object with the specified vocabulary size."""
        self.vocab_size = vocab_size
        self.vocab = None

    def train(self, corpus: List[str]) -> None:
        """
        Train the tokenizer on a corpus of text.

        Args:
            corpus (List[str]): List of strings representing the text corpus.
        """
        # Initialize vocabulary with individual words
        vocab = collections.defaultdict(int)
        for text in corpus:
            for word in text.split():
                vocab[' '.join(word)] += 1

        # Learn merges until reaching vocab_size
        for _ in range(self.vocab_size):
            pairs = self._calculate_pair_stats(vocab)
            if not pairs:
                break
            best_pair = max(pairs, key=pairs.get)
            vocab = self._merge_vocab(best_pair, vocab)

        self.vocab = vocab

    def tokenize(self, text: str) -> List[str]:
        """
        Tokenize the given text based on the learned vocabulary.

        Args:
            text (str): Input text to tokenize.

        Returns:
            List[str]: List of tokens generated from the input text.
        """
        if self.vocab is None:
            raise RuntimeError("Tokenizer has not been trained yet.")

        tokens = []

        for word in text.split():
            symbols = word.split()
            for candidate in symbols:
                if candidate in self.vocab:
                    tokens.append(candidate)
                else:
                    # Tokenize each character individually
                    [tokens.append(char) for char in candidate]
        return tokens

    def _calculate_pair_stats(self, vocab):
        """
        Calculate statistics of symbol pairs in the vocabulary.

        Args:
            vocab: Dictionary containing symbols and their frequencies.

        Returns:
            pairs: Dictionary containing pairs of symbols and their frequencies.
        """
        pairs = collections.defaultdict(int)
        for word, freq in vocab.items():
            symbols = word.split()
            for i in range(len(symbols)-1):
                pairs[(symbols[i], symbols[i+1])] += freq
        return pairs

    def _merge_vocab(self, pair, vocab):
        """
        Merge symbols in the vocabulary based on a given pair.

        Args:
            pair: Pair of symbols to merge.
            vocab: Current vocabulary.

        Returns:
            v_out: Merged vocabulary.
        """
        v_out = {}
        bigram = re.escape(' '.join(pair))
        p = re.compile(r'(?<!\S)' + bigram + r'(?!\S)')
        for word in vocab:
            # Merge the pair of symbols
            w_out = p.sub(''.join(pair), word)
            v_out[w_out] = vocab[word]
        return v_out
```

### Task 2.2. Analysis on Tweets Dataset (10 points)

Train the BPE tokenizer on the tweets dataset. Try to tokenize the tweets with the tokenizer of different `vocab_size`. For example, train the BPE tokenizer with `vocab_size` of [base_vocab_size, 250, 500, 750, 1000]. Plot the dependency of the average length of the tokenized tweet by `vocab_size` to analyze how `vocab_size` affects the length of the tokenized tweet on average. Tell what `vocab_size` is preferrable and why.

```python
import matplotlib.pyplot as plt

def plot_dependency(avg_lengths, vocab_sizes):
    plt.plot(list(avg_lengths.keys()), list(avg_lengths.values()), marker='o')
    plt.xlabel('Vocabulary Size')
    plt.ylabel('Average Tokenized Tweet Length')
    plt.title('Average Tokenized Tweet Length vs Vocabulary Size')
    plt.grid(True)
    plt.xticks(np.arange(0, max(vocab_sizes) + 1, step=100))
    plt.show()
```

```python
import numpy as np
import os

# List of tweets
tweets_dataset = []
file_paths = ["file1", "file2", "file3", "file4", "file5"]

for file_path in file_paths:
    with open(os.path.join("data", file_path), 'r', encoding='utf-8') as file:
        for line in file:
            tweets_dataset.append(line.strip())

# List of vocabulary sizes to experiment with
base_vocab_size = 100
vocab_sizes = [base_vocab_size, 250, 500, 750, 1000]

# Initialize a dictionary to store average tokenized tweet lengths for each vocab size
avg_lengths = {}

# Train the tokenizer for each vocab size and tokenize tweets
for vocab_size in vocab_sizes:
    print("Training BPE on vocab size:", vocab_size)
    tokenizer = BPETokenizer(vocab_size)
    tokenizer.train(tweets_dataset)
    print("Finished training BPE on vocab size:", vocab_size)


    # Tokenize tweets and calculate average tokenized tweet length
    total_length = 0
    for i, tweet in enumerate(tweets_dataset):
        tokens = tokenizer.tokenize(tweet)
        total_length += len(tokens)
    avg_length = total_length / len(tweets_dataset)

    # Store the average length for this vocab size
    avg_lengths[vocab_size] = avg_length

# Plot the dependency of average tokenized tweet length on vocab size
plot_dependency(avg_lengths, vocab_sizes)

# Find the optimal vocab size
optimal_vocab_size = min(avg_lengths, key=avg_lengths.get)
print("Optimal vocabulary size:", optimal_vocab_size)
```

A greater vocabulary size of 1000 is more preferable than lesser sizes because it allows the tokenizer to capture more diverse and specific words and patterns from the tweets, enabling it to tokenize them more accurately and comprehensively, which can improve the understanding and processing of the tweets.
